### T-test to compare category 0 and category 1 for target_b

```python
"""Performing the T-Test: 
For each feature column, an independent two-sample t-test (ttest_ind) is performed to compare the means of the two groups.
Storing and Filtering Results: 
The results (t-statistic and p-values) are stored, converted to a DataFrame, and then filtered to show only those features 
with significant differences (p-value <= 0.05)."""

target_0 = df_target_0.drop(['target_b'], axis=1)
target_1 = df_target_1.drop(['target_b'], axis=1)


import scipy.stats as stats

results = []

# Iterate over each column to perform t-test
for column in target_0.columns:
    # Perform the t-test
    t_stat, p_value = stats.ttest_ind(target_0[column], target_1[column], equal_var=False)
    
    # Append results to the list
    results.append((column, t_stat, p_value))

# Convert results to a DataFrame
results_df = pd.DataFrame(results, columns=['Column', 'T-Statistic', 'P-Value'])
results_df
```

```python
# Filter the results to get only significant columns (p-value <= 0.05)
significant_results = results_df[results_df['P-Value'] <= 0.05]
significant_results
```

####  Why Use a T-Test Before Logistic Regression?

    Initial Filtering:
        You often have many features (predictor variables) in your dataset.
        Not all of these features may be relevant or useful for predicting the outcome.

    Identifying Significant Differences:
        A t-test compares the means of two groups to see if they are significantly different from each other.
        By applying the t-test to each feature, you can determine whether the feature values are significantly different between the two groups of your binary outcome (e.g., target = 0 and target = 1).

    Feature Pre-Selection:
        Features that show significant differences are more likely to be useful for predicting the outcome.
        Pre-selecting features with significant t-test results helps reduce the number of features you need to consider, making the model simpler and more interpretable.
        It helps avoid including irrelevant or noisy features that could negatively impact the performance of your logistic regression model.
        
Understanding the T-Test and P-Value

    Null Hypothesis (H0):
        The null hypothesis for the t-test states that there is no difference between the means of the two groups (i.e., the means are the same).

    Alternative Hypothesis (H alfa):
        The alternative hypothesis states that there is a difference between the means of the two groups (i.e., the means are not the same).

    P-Value:
        The p-value measures the probability of obtaining a t-test statistic as extreme as, or more extreme than, the one observed, assuming that the null hypothesis is true.
        In simpler terms, it indicates the likelihood that the observed difference in means happened by chance.

Interpreting the P-Value

    Low P-Value (â‰¤ 0.05):
        Indicates strong evidence against the null hypothesis.
        Suggests that the difference in means is statistically significant and unlikely to have occurred by random chance.
        You reject the null hypothesis and conclude that the means are different.

    High P-Value (> 0.05):
        Indicates weak evidence against the null hypothesis.
        Suggests that any observed difference in means could easily have occurred by random chance.
        You fail to reject the null hypothesis and conclude that there is no significant difference in means.

Key Points

    P-Value: Probability of observing the data (or something more extreme) assuming the null hypothesis is true.
    Low P-Value (â‰¤ 0.05): Reject the null hypothesis, significant difference in means.
    High P-Value (> 0.05): Fail to reject the null hypothesis, no significant difference in means.

#### Why It Matters for Feature Selection

By using the p-value from the t-test, you can pre-select features that show significant differences between your target groups. This helps you focus on features that are more likely to be relevant for predicting the outcome, improving the performance and interpretability of your logistic regression model.

#### What is Logistic Regression?

Logistic regression is a type of statistical model used to predict the probability of a binary outcome (like yes/no, true/false, 0/1) based on one or more predictor variables (features). Instead of predicting the outcome directly, logistic regression predicts the odds of the outcome occurring.

How Does It Work?

    Log Odds:
        The model calculates the log odds of the outcome happening. The log odds is just the natural logarithm of the odds, where the odds are the probability of the outcome occurring divided by the probability of it not occurring.

    Coefficients:

        These coefficients tell you how much the log odds of the outcome change with a one-unit increase in the predictor variable.

    Odds Ratios:

Interpreting P-values

#### Random Forest

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import recall_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

y_train_SMOTE = train_smote['target_b']
X_train_SMOTE = train_smote.drop(['target_b'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X_train_SMOTE, y_train_SMOTE, test_size=0.2, random_state=42)

max_depth = range(1, 50)
test_recall = []

for depth in max_depth:
    model_tree = DecisionTreeClassifier(max_depth=depth, random_state=42)
    model_tree.fit(X_train, y_train)
    y_pred = model_tree.predict(X_test)
    recall = recall_score(y_test, y_pred)  # Calculate recall score
    test_recall.append(recall)

# Plotting the recall scores
plt.figure(figsize=(10, 6))
plt.plot(max_depth, test_recall, marker='o', linestyle='-', color='b', label='Test Recall')
plt.title('Decision Tree Classifier Performance')
plt.xlabel('Maximum Depth')
plt.ylabel('Recall')
plt.xticks(max_depth)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

y_train_SMOTE = train_smote['target_b']
X_train_SMOTE = train_smote.drop(['target_b'], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X_train_SMOTE, y_train_SMOTE, test_size=0.2, random_state=42)

#To check the best set of parameters returned
n_estimators_list = [50,100]
max_depth_list = [24,31,34]

auc_list =[]

for n_estimators in n_estimators_list:
    for max_depth in max_depth_list:
        # Initialize and train RandomForestClassifier
        model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
        model.fit(X_train, y_train)
        
        # Predict probabilities and calculate ROC-AUC
        y_pred_proba = model.predict_proba(X_test)[:, 1]  # Probability of class 1
        fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_proba)
        auc = metrics.auc(fpr, tpr)
        
        # Append AUC score to list
        auc_list.append((n_estimators, max_depth, auc))

# Print the list of AUC scores for each parameter combination
for auc_result in auc_list:
    print(f"n_estimators = {auc_result[0]}, max_depth = {auc_result[1]}, AUC = {auc_result[2]}")
```
